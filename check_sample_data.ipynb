```python
from scipy.io.wavfile import read

wavfile = "/Users/seiji/data/fan/id_00/normal/00000405.wav"
anowavfile = "/Users/seiji/data/fan/id_00/abnormal/00000162.wav"

fs, data = read(wavfile)
anofs, anodata = read(anowavfile)

print ("Sampling rate :", fs)
data = data[:,0]
anodata = anodata[:,0]
print(data.shape)
```

```python
from scipy import signal
from scipy.fft import fftshift
import matplotlib.pyplot as plt

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = 10
fs = 16000
print("normal")
f, t, Sxx = signal.spectrogram(data, fs)
plt.pcolormesh(t, f,10*np.log( Sxx ), shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
print("anormal")
f, t, Sxx = signal.spectrogram(anodata, fs)
plt.pcolormesh(t, f,10*np.log( Sxx ), shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
